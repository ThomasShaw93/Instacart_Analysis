```python
#Create product ranges
product_groups = []

for value in df_products['prices']:
    if (value < 5):
        product_groups.append('Low-range Product')
    elif (value >= 6) and (value <= 15):
        product_groups.append('Mid-range Product')
    else:
        product_groups.append('High-range Product')
```

```python
# Create new column in products dataframe
df_products['price_range'] = product_groups
```

```python
# View products dataframe
df_products
```

```python
product_pie_chart = df_products['price_range'].value_counts().sort_index().plot.pie(ylabel = 'Price range', autopct = '%1.1f%%')
```

```python
# Exporting pie chart
product_pie_chart.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'product_pie_chart.png'))
```
