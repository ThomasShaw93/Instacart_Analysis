# 03. Deriving New Variables

```python
# Create subset
df = ords_prods_merge[:1000000]
```

```python
# Define a Function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Check most expensive price
df['prices'].max()
```

```python
# Predefined function 1
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

# 04. Task

```python
# 2) Loop for busiest/least busy days of the week 
results = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value in [0,1]:
    results.append("Busiest days")
  elif value in [3,4]:
    results.append("Slowest days")
  else:
    results.append("Regularly busy")
```

```python
# Add busiest days column to dataframe
ords_prods_merge['busiest_days'] = results
```

```python
# Value count
ords_prods_merge['busiest_days'].value_counts()
```

3) Busiest days now equals the sum of Saturdays (0) and Sundays (1). Slowest days equals the sum of Tuesdays (3) and Wednesdays (4). The number of regularly busy days has reduced.

```python
# 4) busiest time of day value count
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

From the above we can see that sequential hours of 0-6 are when the lowest hours occur. On the other end of the spectrum, the sequential hours of 9-16 are when the most orders are made. These will be my two grouped time periods, with the remaining hours being 'average orders'.

```python
# 2) Loop for busiest/least hours of day
result1 = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value >= 9 and value <= 16:
    result1.append("Most Orders")
  elif value >= 0 and value <=6:
    result1.append("Fewest Orders")
  else:
    result1.append("Average Orders")
```

```python
# Add busiest period of day column to dataframe
ords_prods_merge['busiest_period_of_day'] = result1
```

```python
# Print frequency
ords_prods_merge['busiest_period_of_day'].value_counts()
```
